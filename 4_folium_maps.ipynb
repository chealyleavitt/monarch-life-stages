```python
# Import Libraries
import pandas as pd
import geopandas
import folium
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('class_predictions_2019_batch_1.csv')
df2 = pd.read_csv('class_predictions_2019_batch_2.csv')
df3 = pd.read_csv('class_predictions_2019_batch_3.csv')

predictions = df1.append(df2).append(df3)
```

```python
len(predictions)
```

```python
predictions['obs_id'] = [int(x[:-4]) for x in predictions.filename]
```

```python
predictions.columns
```

```python
observations = pd.read_csv('observations-285599.csv')
```

```python
obs = observations[['id','observed_on',
                    'latitude','longitude',
                    'species_guess','image_url','url']]
pred = predictions[['obs_id','prediction','score']]
```

```python
pred = pd.merge(pred, obs, left_on = 'obs_id', right_on = 'id', how = 'left')
```

```python
geometry = geopandas.points_from_xy(pred.longitude, pred.latitude)
geo_df = geopandas.GeoDataFrame(pred[['obs_id','prediction','latitude','longitude']], geometry = geometry)
geo_df.head()
```

```python
repro = geo_df[geo_df.prediction!='adult']

from folium import plugins

map = folium.Map(location=[43.6747, -72.158], tiles="Stamen Toner", zoom_start=7)

heat_data = [[point.xy[1][0], point.xy[0][0]] for point in repro.geometry]

heat_data
plugins.HeatMap(heat_data, radius = 15).add_to(map)
map
```
